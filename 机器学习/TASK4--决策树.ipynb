```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

%matplotlib inline
warnings.filterwarnings('ignore')
plt.rcParams['font.sans-serif']=['SimHei']
```

```python
data = pd.read_csv('G:/CDO/狗熊会机器学习/TASK4--刘嘉玲/Speed Dating Data.csv', encoding='gbk')
data.head()
```

# 任务一

```python
figure = plt.figure(1, (12, 5))  # 设置画布

data_reject = data[data['dec']==0]  # 最后结果为接受对象
data_accept = data[data['dec']==1]  # 最后结果为拒绝对象

# 是否接收对象与吸引力的箱图
ax1 = plt.subplot(131)  # 1*3分布的画图中的第一个子图

# boxplot不接受缺失值Nan，因此先判断 'attr' 属性为不为空的行
ax1.boxplot(data_reject.loc[data_reject['attr'].notnull(), 'attr'], sym='', positions=[0], widths=[0.5], labels=['拒绝对象'])
ax1.boxplot(data_accept.loc[data_accept['attr'].notnull(), 'attr'], sym='', positions=[1], widths=[0.5], labels=['接受对象'])
ax1.set_title('吸引力对相亲决策的影响')  # 设置标题
ax1.set_ylabel('对方的吸引力得分')  # 设置纵坐标
ax1.set_xlabel('决策：是否接收相亲对象')  # 设置横坐标

# 是否接收对象与真诚得分的箱图
ax2 = plt.subplot(132)
ax2.boxplot(data_reject.loc[data_reject['sinc'].notnull(), 'sinc'],  sym='', positions=[0], widths=[0.5], labels=['拒绝对象'])
ax2.boxplot(data_accept.loc[data_accept['sinc'].notnull(), 'sinc'],  sym='', positions=[1], widths=[0.5], labels=['接受对象'])
ax2.set_title('真诚对相亲决策的影响')
ax2.set_ylabel('对方的真诚得分')
ax2.set_xlabel('决策：是否接收相亲对象')

# 是否接受对象与智商得分的箱图
ax3 = plt.subplot(133)
ax3.boxplot(data_reject.loc[data_reject['intel'].notnull(), 'intel'], sym='', positions=[0], widths=[0.5], labels=['拒绝对象'])
ax3.boxplot(data_accept.loc[data_accept['intel'].notnull(), 'intel'], sym='', positions=[1], widths=[0.5], labels=['接受对象'])
ax3.set_title('智商对相亲决策的影响')
ax3.set_ylabel('对方的智商得分')
ax3.set_xlabel('决策：是否接收相亲对象')
plt.show()
```

# 任务二

```python
#女性
columns=['dec','gender','attr','sinc','intel','fun','amb','shar']
column_dict = {'dec':'决定', 'gender':'性别', 'attr':'吸引力', 'sinc':'真诚', 
               'intel':'智商', 'fun':'幽默', 'amb':'抱负', 'shar':'共同爱好'}

data_4_1 = data[columns]
female = data_4_1[data_4_1['gender']==0] 
```

```python
#男性
columns=['dec','gender','attr','sinc','intel','fun','amb','shar']
column_dict = {'dec':'决定', 'gender':'性别', 'attr':'吸引力', 'sinc':'真诚', 
               'intel':'智商', 'fun':'幽默', 'amb':'抱负', 'shar':'共同爱好'}

data_4_1 = data[columns]
male = data_4_1[data_4_1['gender']==1] 
```

# 任务三

```python
from sklearn import tree
import graphviz 
import pydotplus
import os
# min_sample_split表示划分节点的最小样本数量
# max_depth表示树的最大深度
# min_impurity_decrease表示继续划分的不纯度下降阈值
# min_impurity_split表示当前节点划分的不纯度阈值
# 可以调整参数观察树形状的改变 
```

```python
model_female = tree.DecisionTreeClassifier(min_samples_split=50, max_depth=3, 
                                           min_impurity_split=0.35)
female_discard_missing = female.loc[np.all(female.notnull(), axis=1), :]
model_female.fit(female_discard_missing[female.columns[2:]], female_discard_missing['dec']) 
```

```python
dot_tree_female = tree.export_graphviz(model_female, out_file=None, 
                                       feature_names=female.columns[2:],
                                       class_names=['No', 'Yes'],
                                       filled=True, rounded=True, 
                                       special_characters=True)
graph_female = pydotplus.graph_from_dot_data(dot_tree_female)
graph_female.write_png("G:/CDO/狗熊会机器学习/TASK4--刘嘉玲/female_decision_tree.png")  # 保存图片
```

```python
model_male = tree.DecisionTreeClassifier(min_samples_split=50, max_depth=3, 
                                           min_impurity_split=0.35)
male_discard_missing = male.loc[np.all(male.notnull(), axis=1), :]
model_male.fit(male_discard_missing[female.columns[2:]], male_discard_missing['dec'])
```

```python
dot_tree_male = tree.export_graphviz(model_male, out_file=None, 
                                       feature_names=male.columns[2:],
                                       class_names=['No', 'Yes'],
                                       filled=True, rounded=True, 
                                       special_characters=True)
graph_male = pydotplus.graph_from_dot_data(dot_tree_male)
graph_male.write_png("G:/CDO/狗熊会机器学习/TASK4--刘嘉玲/male_decision_tree.png")  # 保存图片
```
