**‘x_num_0’，‘x_num_1’和‘x_num_3’中存在-99，与其他值差距很大，认为是缺失值，需要处理**

```python
data_b_train=data_b_train.replace(-99,np.nan)
```

```python
data_b_train.info()
```

```python
data_b_train.isnull().sum()
```

```python
#使用随机森林对缺失值进行预测填充
from sklearn.ensemble import RandomForestRegressor

#把要填充的特征和其他无缺失的特征取出，这里先处理x_num_0特征
qs_columns=['x_num_0','x_num_1','x_num_3'] #有缺失的列名
for i in range(len(qs_columns)):
    data1_add=data_b_train[qs_columns[i]]
    data1=data_b_train[['x_num_2','x_num_4','x_cat_0','x_cat_1','x_cat_2','x_cat_3','x_cat_4','x_cat_5','x_cat_6','x_cat_7','x_cat_8','x_cat_9','x_cat_10','x_cat_11','x_cat_12']]
    data1.insert(0,qs_columns[i],data1_add)


    #成缺失值和非缺失值两部分，分别作为训练集和测试集
    know=data1[data1[qs_columns[i]].notnull()].values
    unknow=data1[data1[qs_columns[i]].isnull()].values

    #获取训练集特征和结果标签
    X=know[:,1:] #训练集特征
    y=know[:,0]  #训练集的结果标签

    #利用上面构建的训练集训练随机森林回归模型
    RF_clf=RandomForestRegressor(random_state=0,n_estimators=200,n_jobs=-1)
    RF_clf.fit(X,y)

    #用得到的模型对年龄位置的样本进行预测
    y_predicted=RF_clf.predict(unknow[:,1:])

    #用得到的预测结果填补原缺失数据
    data_b_train.loc[(data1[qs_columns[i]].isnull()),qs_columns[i]]=y_predicted
```

```python
#'x_num_8','x_num_55'没有一个数据，需要删除
#‘x_num_18’,‘x_num_43'中绝大部分都是缺失值，直接删除
data_m_train=data_m_train.drop(columns=['x_num_8','x_num_18','x_num_43','x_num_55'])
```

```python
#'x_num_41'存在少量缺失值，需填充

import numpy as np
#成缺失值和非缺失值两部分，分别作为训练集和测试集
know_m=data_m_train[data_m_train['x_num_41'].notnull()].values
unknow_m=data_m_train[data_m_train['x_num_41'].isnull()].values

#获取训练集特征和结果标签
X_m=np.c_[know_m[:,:37],know_m[:,38:]] #训练集特征
y_m=know_m[:,37]  #训练集的结果标签

#利用上面构建的训练集训练随机森林回归模型
RF_clf=RandomForestRegressor(random_state=0,n_estimators=200,n_jobs=-1)
RF_clf.fit(X_m,y_m)

#用得到的模型对年龄位置的样本进行预测
y_predicted=RF_clf.predict(np.c_[unknow_m[:,:37],unknow_m[:,38:]])

#用得到的预测结果填补原缺失数据
data_m_train.loc[(data_m_train['x_num_41'].isnull()),'x_num_41']=y_predicted
```

**XGBoost模型参数调优**

这里设置固定的学习速率 learning rate为0.3，n_estimators=100，也可以对其进行调优

```python
#max_depth和min_child_weight参数调优
#max_depth和min_child_weight参数对最终结果有很大的影响。max_depth通常在1-10之间，min_child_weight。采用栅格搜索（grid search），我们先大范围地粗略参数，然后再小范围的微调。
#网格搜索scoring = 'roc_auc' 只支持二分类，多分类需要修改scoring（默认支持多分类）

param_test1 = {
'max_depth':[i for i in range(1,10,1)],
'min_child_weight':[i for i in range(1,6,1)]
}
from sklearn import svm, model_selection, datasets
gsearch = model_selection.GridSearchCV(
estimator = XGBClassifier(
learning_rate =0.3,
n_estimators=100, max_depth=5,
min_child_weight=1,
gamma=0,
subsample=0.8,
colsample_bytree=0.8,
objective= 'binary:logistic',
nthread=4,
scale_pos_weight=1,
seed=27),
param_grid = param_test1,
scoring='roc_auc',
n_jobs=4,
cv=5)
gsearch.fit(X_train,y_train)

print('max_depth_min_child_weight')
#print('gsearch1.grid_scores_', gsearch.cv_results_)
print('gsearch1.best_params_', gsearch.best_params_)
print('gsearch1.best_score_', gsearch.best_score_)
```

```python
# gamma参数调优
# 在已经调整好其他参数的基础上，进行gamma参数的调优了。Gamma参数取值范围很大，这里设置为5，其实你也可以取更精确的gamma值。

param_test3 = {
 'gamma':[i/10.0 for i in range(0,5)]
}
gsearch = model_selection.GridSearchCV(
estimator = XGBClassifier(
learning_rate =0.1,
n_estimators=140,
max_depth=4,
min_child_weight=5,
gamma=0,
subsample=0.8,
colsample_bytree=0.8,
objective= 'binary:logistic',
nthread=4,
scale_pos_weight=1,
seed=27),
param_grid = param_test3,
scoring='roc_auc',
n_jobs=4,
cv=5)
gsearch.fit(X_train,y_train)
print('gamma')
#print('gsearch1.grid_scores_', gsearch.cv_results_)
print('gsearch1.best_params_', gsearch.best_params_)
print('gsearch1.best_score_', gsearch.best_score_)
```

```python
#调整subsample 和 colsample_bytree参数
#尝试不同的subsample 和 colsample_bytree 参数。分两个阶段来进行这个步骤。这两个步骤都取0.6,0.7,0.8,0.9作为起始值。
#取0.6,0.7,0.8,0.9作为起始值

param_test4 = {
 'subsample':[i/10.0 for i in range(6,10)],
 'colsample_bytree':[i/10.0 for i in range(6,10)]
}

gsearch = model_selection.GridSearchCV(
estimator = XGBClassifier(
learning_rate =0.1,
n_estimators=177,
max_depth=4,
min_child_weight=5,
gamma=0.4,
subsample=0.8,
colsample_bytree=0.8,
objective= 'binary:logistic',
nthread=4,
scale_pos_weight=1,
seed=27),
param_grid = param_test4,
scoring='roc_auc',
n_jobs=4,
cv=5)
gsearch.fit(X_train,y_train)
print('subsample_colsample_bytree------------------')
#print('gsearch1.grid_scores_', gsearch.cv_results_)
print('gsearch1.best_params_', gsearch.best_params_)
print('gsearch1.best_score_', gsearch.best_score_)
```

```python
#正则化参数调优reg_alpha
#由于gamma函数提供了一种更加有效的降低过拟合的方法，大部分人很少会用到这个参数，但是我们可以尝试用一下这个参数。

param_test6 = {
 'reg_alpha':[1e-5, 1e-2, 0.1, 1, 100]
}

gsearch = model_selection.GridSearchCV(
estimator = XGBClassifier(
learning_rate =0.1,
n_estimators=177,
max_depth=4,
min_child_weight=5,
gamma=0.4,
subsample=0.9,
colsample_bytree=0.8,
objective= 'binary:logistic',
nthread=4,
scale_pos_weight=1,
seed=27),
param_grid = param_test6,
scoring='roc_auc',
n_jobs=4,
cv=5)
gsearch.fit(X_train,y_train)
print('reg_alpha------------------')
print('gsearch1.grid_scores_', gsearch.cv_results_)
print('gsearch1.best_params_', gsearch.best_params_)
print('gsearch1.best_score_', gsearch.best_score_)
```

```python
#正则化参数调优reg_lambda
#由于gamma函数提供了一种更加有效的降低过拟合的方法，大部分人很少会用到这个参数，但是我们可以尝试用一下这个参数。

param_test7 = {
 'reg_lambda':[1e-5, 1e-2, 0.1, 1, 100]
}
gsearch = model_selection.GridSearchCV(
estimator = XGBClassifier(
learning_rate =0.1,
n_estimators=177,
max_depth=4,
min_child_weight=5,
gamma=0.4,
subsample=0.9,
colsample_bytree=0.8,
objective= 'binary:logistic',
nthread=4,
scale_pos_weight=1,
seed=27),
param_grid = param_test7,
scoring='roc_auc',
n_jobs=4,
cv=5)
gsearch.fit(X_train,y_train)
print('reg_lambda------------------')
#print('gsearch1.grid_scores_', gsearch.cv_results_)
print('gsearch1.best_params_', gsearch.best_params_)
print('gsearch1.best_score_', gsearch.best_score_)
```
